# Two Sum

ðŸ”¢ The Problem (Plain English)

You are given:
- A list of numbers
- A target number

Your job:

- Find two different numbers in the list
- Whose sum equals the target
- Return their indices (positions)

Example 1 :-
- nums = [2, 7, 11, 15]
- target = 9

Example 2 :-
- nums = [4, 1, 5, 3]
- target = 6

```python
# The below given solution is the brute force solution
def two_sum(nums, target):
    matching_pair_index = []
    len_nums = len(nums)
    for i, num in enumerate(nums):
        for j in range(i+1, len_nums):
            print(nums[i], nums[j])
            if nums[i] + nums[j] == target:
                matching_pair_index.append((i, j))
    return matching_pair_index
```

```python
nums = [2,7,11,3,15,6]
target = 9
two_sum(nums, target)
```

```python
# Now Lets write the smart way, hash map solution
def two_sum_hash_map(nums, target):
    seen = {}
    matched_pair_index = []
    for i, num in enumerate(nums):
        diff = target - num
        if diff in seen:
            matched_pair_index.append((seen[diff], i))
            # return [seen[diff], i]
        seen[num] = i
        print(f"seen: {seen}")
    return matched_pair_index
nums = [2,7,11,3,15,6]
target = 9
two_sum_hash_map(nums, target)
```

```python
nums = [4, 1, 5, 3]
target = 6
two_sum_hash_map(nums, target)
```

# Sliding Window

Problem: Longest Substring Without Repeating Characters

The Question:- Given a string s, find the length of the longest substring without repeating characters.

Important:

- Substring means continuous.
- No duplicate characters allowed inside it.

### **Example 1**

s = "abcabcbb"

What is the longest substring without repeating characters?

Letâ€™s scan:

"abc" â†’ valid (length 3)

Next comes another "a" â†’ duplicate

Then "bca" â†’ valid but length 3

Then "cab" â†’ 3

Then "abc" â†’ 3

Then duplicates start dominating

So answer = 3

```python
# The below given solution is the brute force solution
def longest_substring_without_repeating_characters(s):
    longest_substring = []
    for i, char in enumerate(s):
        for j in range(i+1, len(s)):
            if s[i] == s[j]:
                longest_substring.append(s[i:j])
                break
    return len(max(longest_substring, key=len))
```

```python
s = "abba"
longest_substring_without_repeating_characters(s)
```

```python
# The below given solution is the hash map solution
def lengthOfLongestSubstring(s):
    seen = set()
    l = 0
    max_len = 0
    
    for r in range(len(s)):
        print(f"s[r]: {s[r]} and Seen : {seen} & l : {l} & max_len : {max_len} & r : {r}")
        while s[r] in seen:
            seen.remove(s[l])
            l += 1
        seen.add(s[r])
        max_len = max(max_len, r - l + 1)
    
    return max_len

s = "abcabcbb"
lengthOfLongestSubstring(s)
```

# Valid Anagram

Two strings are given. Check if they are anagrams.

Anagram = same letters, different order.

Example 1 ;-
s = "anagram"
t = "nagaram"


Same letters â†’ True

```python

def isAnagram(s, t):
    return sorted(s) == sorted(t)

s = "anagram"
t = 'nagaram'
isAnagram(s, t)
```

```python
from collections import Counter

# The below given solution is the hash map solution
def isAnagram(s, t):
    return Counter(s) == Counter(t)

s = 'anagram'
t = 'nagaram'
isAnagram(s, t)


```

# Product of Array Except Self

This one confuses many people.

The Question

Given:

nums = [1,2,3,4]

Return : [24,12,8,6]

```python
from typing import List

def product_of_array_except_self(nums:List[int]) -> List[int]:
    product_array = []
    for i, num1 in enumerate(nums):
        product = 1
        for j, num2 in enumerate(nums):
            if i != j:
                product = product * num2
        product_array.append(product)
    return product_array

nums = [1, 2, 0, 4]
product_of_array_except_self(nums)
```

```python
# The below given solution is the hash map solution
from typing import List

def product_of_array_except_self(nums: List[int]) -> List[int]:
    n = len(nums)
    res = [1] * n
    print(f"res: {res}")

    # prefix products
    prefix = 1
    for i in range(n):
        res[i] = prefix
        prefix *= nums[i]
        print(f"res: {res} & prefix: {prefix} & i: {i} & nums[i]: {nums[i]}")

    # suffix products
    suffix = 1
    for i in range(n - 1, -1, -1):
        res[i] *= suffix
        suffix *= nums[i]
        print(f"res: {res} & suffix: {suffix}")

    return res

nums = [1,2,3,4]
product_of_array_except_self(nums)
```
